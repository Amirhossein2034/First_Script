```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris_sklearn = load_iris()
iris_df = pd.DataFrame(iris_sklearn.data, columns=iris_sklearn.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris_sklearn.target, iris_sklearn.target_names)

# Clean the column names for easier access
iris_df.columns = [col.replace(" (cm)", "").replace(" ", "_") for col in iris_df.columns]

# Create a scatter plot: Sepal Length vs Petal Length
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=iris_df,
    x="sepal_length",
    y="petal_length",
    hue="species",
    style="species",
    palette="Set1"
)
plt.title("Sepal Length vs Petal Length by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

# Load the Iris dataset
iris_sklearn = load_iris()
X = iris_sklearn.data
y = iris_sklearn.target
target_names = iris_sklearn.target_names

# Apply PCA to reduce dimensions to 2
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Create a DataFrame with principal components and species labels
iris_pca_df = pd.DataFrame(data=X_pca, columns=["PC1", "PC2"])
iris_pca_df["species"] = pd.Categorical.from_codes(y, target_names)

# Plot the PCA-reduced data
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=iris_pca_df,
    x="PC1",
    y="PC2",
    hue="species",
    style="species",
    palette="Set2"
)
plt.title("PCA of Iris Dataset (2 Components)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend(title="Species")
plt.grid(True)
plt.tight_layout()
plt.show()
```
